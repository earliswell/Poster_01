```python
def convert_exit_number(exit_num):
    # 먼저 exit_num을 문자열로 변환
    exit_num = str(exit_num)
    
    if '-' in exit_num:  # '2-1'과 같은 경우
        return float(exit_num.replace('-', '.'))
    else:  # 그 외의 경우
        return float(exit_num)

# 출구 번호를 숫자와 문자열로 나누어 처리
df_concat['출구번호_변환'] = df_concat['출구번호'].apply(convert_exit_number)
```

```python
import requests
import pandas as pd

def get_exit_coordinates_kakao(station_name, api_key):
    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={station_name}'
    headers = {"Authorization": f"KakaoAK {api_key}"}
    
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        results = response.json()['documents']
        if results:  # 결과가 있을 경우
            # 첫 번째 결과만 반환 (가장 관련성 높은 결과)
            result = results[0]
            return {
                '출구 이름': result['place_name'],
                '위도': result['y'],
                '경도': result['x']
            }
    return None  # 결과가 없을 경우

# API 키 설정
api_key = 'efc3b0abd5dd7a662705dc3b1f7091ba'

# 결과를 저장할 리스트 초기화
exit_data = []

# df_concat의 각 출구에 대해 좌표 정보 가져오기
for search_term in df_concat['search']:
    exit_info = get_exit_coordinates_kakao(search_term, api_key)
    if exit_info:
        exit_data.append(exit_info)

# exit_data를 데이터프레임으로 변환
df_exit_info = pd.DataFrame(exit_data)

# 새로운 데이터프레임 저장
# df_exit_info.to_csv('exit_coordinates.csv', index=False)

df_exit_info.head()
```
