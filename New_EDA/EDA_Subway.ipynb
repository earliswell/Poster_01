```python
df_4 = pd.read_excel('./data/수도권9호선주소.xlsx')
df_4.head()
```

```python
df_4.fillna(1, inplace=True)
```

```python
df_4.columns = ['철도운영기관명', '선명', '역명', '지번주소', '역사도로명주소', '출구번호']
```

```python
df_4 = df_4[['철도운영기관명', '선명', '역명', '출구번호', '역사도로명주소']]
```

```python
df_4['출구번호'] = df_4['출구번호'].astype('int')
```

```python
df_4.columns.tolist()
```

```python
def replicate_rows(row):
    return pd.DataFrame([row] * row['출구번호'])

# 각 행을 출구번호에 따라 복제
result = pd.concat([replicate_rows(row) for _, row in df_4.iterrows()], ignore_index=True)
result['출구번호'] = result.groupby(['철도운영기관명', '선명', '역명', '역사도로명주소']).cumcount() + 1

result.head()
```

```python
def convert_exit_number(exit_num):
    if pd.isna(exit_num):  # NaN 값인 경우
        return np.nan
    if '-' in exit_num:  # '2-1'과 같은 경우
        return float(exit_num.replace('-', '.'))
    else:  # 그 외의 경우
        return float(exit_num)

# 출구 번호를 숫자와 문자열로 나누어 처리
df_sub['출구번호_변환'] = df_sub['출구번호'].apply(convert_exit_number)
```

```python
df_sub.sort_values(by=['선명', '역명', '출구번호_변환'], inplace=True)
```

```python
df_sub.reset_index(drop=True, inplace=True)
```
