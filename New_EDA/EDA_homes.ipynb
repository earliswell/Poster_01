```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/multi.csv')
df.head()
```

```python
df['yyyymmdd'] = pd.to_datetime(df['yyyymmdd'], format="%Y%m%d")
```

```python
df['year'] = df['yyyymmdd'].dt.year
df['quarter'] = df['yyyymmdd'].dt.quarter
```

```python
df.sort_values(ascending=True, by='yyyymmdd', inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('./data/home.csv', index=False)
```
