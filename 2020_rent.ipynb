```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
import numpy as np
```

```python
rent_2020 = pd.read_csv('./체류거점/체류거점_2020.csv')
df = pd.read_csv('./체류거점/스타벅스_집계구.csv')

print(rent_2020.shape, df.shape)
rent_2020.head()
```

```python
df['geometry'] = df['geometry'].apply(wkt.loads)

gdf = gpd.GeoDataFrame(df, geometry='geometry')

gdf.head()
```

```python
gdf.columns
```

```python
gdf.drop(['index_right', '인허가일자', '영업상태명', '폐업일자', '사업장명', '좌표정보(x)', '좌표정보(y)', 'Year', 'Latitude', 'Longitude'], axis=1, inplace=True)
```

```python
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(rent_2020['longitude'], rent_2020['latitude'])]
rent_gdf = gpd.GeoDataFrame(rent_2020, geometry=geometry, crs='EPSG:4326')

rent_gdf
```

```python
gdf.set_crs(epsg=4326, inplace=True)
gdf.crs
```

```python
gpd.sjoin(gdf, rent_gdf)
```
